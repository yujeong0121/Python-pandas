#### pd.crosstab() - titanic

```python
titanic = pd.read_csv("./Course_Materials_Part2/Course_Materials_Part2/Video_Lecture_NBs/titanic.csv")
```

```python
titanic.head()
```

```python
titanic.head()
```

```python
titanic.info()
```

```python
# 성별과 객실 조합 합계
pd.crosstab(titanic.sex, titanic.pclass, margins=True)
```

```python
# normaliz는 상대값을 구한다. 891명의 승객 중에서 1등실을 이용한 여성 승객은 총 94명인데 이 숫자를 퍼센트로 표시
# 총 승객 대비 약 10.5%
pd.crosstab(titanic.sex, titanic.pclass, margins=True, normalize= True)
```

```python
titanic.groupby(["sex", "pclass"]).pclass.count().unstack()
```

crosstab은 서로 다른 데이터 프레임에서 가져온 열을 사용할 수 있는 장점이 있다. groupby는 같은 데이터 프레임에서만 사용 가능하다.

```python
pd.crosstab(index=titanic.sex, columns=titanic.pclass, values=titanic.age,aggfunc="mean")
# titanic.pivot_table(index = "sex", columns="pclass", values = "age", aggfunc="mean")와 같은 결과
# titanic.groupby(["sex", "pclass"]).age.mean().unstack()와 같은 결과
# crosstab이 확장성이 더 좋음, 하나의 데이터 프레임에 국한되지 않디 때문
```

```python
pd.crosstab(index=titanic.sex, columns=titanic.pclass, values=titanic.fare,aggfunc="sum", margins=True, normalize=True)
```

#### pd.crosstab() - summer

```python
summer = pd.read_csv("./Course_Materials_Part2/Course_Materials_Part2/Video_Lecture_NBs/summer.csv")
```

```python
summer.head()
```

```python
pd.crosstab(index=[summer.Year, summer.Country], columns=summer.Medal, aggfunc= "count", values = summer.Athlete, margins=True).fillna(0)
```

```python
summer.pivot_table(index=["Year","Country"], columns="Medal", aggfunc= "count", values = "Athlete", fill_value=0, margins=True)
```

```python
summer.groupby(["Year", "Country", "Medal"]).Medal.count().unstack(fill_value = 0)
```

여러가지 방법으로 피봇 테이블을 만들 수 있다. 단 groupby는 합계를 구하는 옵션이 없어서 몇 가지 작업을 추가로 해줘야 한다.

#### Melting DataFrames with melt() - pivot의 반대

```python
table_2012 = pd.read_csv("./Course_Materials_Part2/Course_Materials_Part2/Video_Lecture_NBs/table_2012.csv")
```

```python
table_2012
```

```python
table_2012.shape
```

```python
table_2012.melt(id_vars="Country", value_vars=["Gold", "Silver", "Bronze"], var_name = "Medal", value_name="Count") # 하나의 열로 합치기
```

```python
melt_2012 = table_2012.melt(id_vars="Country", value_vars=["Gold", "Silver", "Bronze"], var_name = "Medal", value_name="Count") # 하나의 열로 합치기
```

```python
melt_2012.shape
```

```python
melt_2012
```

```python
# 원래 형태로 되돌리기
melt_2012.pivot(index="Country", columns = "Medal", values = "Count").sort_values("Gold", ascending =False)
```

pivot 메소드는 데이터 프레임을 롱 포맷에서 와이드 포맷으로 바꾸는 기능을 하고, melt 메소드는 데이터 프레임을 언피벗해서 와이드 포맷을 롱 포맷으로 바꾸는 기능을 한다.
