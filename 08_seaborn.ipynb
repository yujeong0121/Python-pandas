### Advanced / statistical plotting with Seaborn

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic = pd.read_csv("./Course_Materials_Part2/Course_Materials_Part2/Video_Lecture_NBs/titanic.csv")
```

```python
titanic.head()
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=2, palette= "viridis") # 미적 매개변수
sns.countplot(data = titanic, x = "sex", hue = "pclass") # hue 범주형 데이터가 있는 열 지정 가능
plt.show()
```

#### Categorical Plots

```python
# stripplot(산포도)
# jitter:어디에 밀집되어 있는지 잘 보기 위해 점을 겹치기 않고 표현
# dodge: 범주별로 점 분리
plt.figure(figsize=(12,8))
sns.set(font_scale=1.5)
sns.stripplot(data = titanic, x = "sex", y = "age", jitter = True, hue = "pclass", dodge = True)
plt.show()
```

```python
# swarmplot
# 같은 항목을 나란히 배열하기 때문에 분포를 더 쉽게 확인
# 3등실 남성 승객들 중에 가장 많은 수의 나이대가 분포한 부분이 잘 보임
# 나머지 등급은 나이 분포가 균일함
plt.figure(figsize=(12,8))
sns.set(font_scale=1.5)
sns.swarmplot(data = titanic, x = "sex", y = "age", hue = "pclass", dodge = True)
plt.show()
```

```python
# violinplot(선과 곡선으로 나타내는 그래프)
# 남성: 1등급-40대, 2등급 - 30대, 3등급 - 20대가 많다
plt.figure(figsize=(12,8))
sns.set(font_scale=1.5)
sns.violinplot(data = titanic, x = "sex", y = "age", hue = "pclass", dodge = True)
sns.swarmplot(data = titanic, x = "sex", y = "age", hue = "pclass", dodge = True, color = "black")
plt.show()
```

```python
# split = False
plt.figure(figsize=(12,8))
sns.set(font_scale=1.5)
sns.violinplot(data = titanic, x = "pclass", y = "age", hue = "sex", dodge = True, split = False )
plt.show()
```

```python
# split = True
plt.figure(figsize=(12,8))
sns.set(font_scale=1.5)
sns.violinplot(data = titanic, x = "pclass", y = "age", hue = "sex", dodge = True, split = True )
plt.show()
```

```python
# barplot(막대그래프)
# 통계적인 그래프, 검은선은 신뢰구간, 선의 가운데는 평균
# 타이타닉 호의 승선 예약을 1000번 반복해서 시뮬레이션 한 결과를 토대로 95%의 신뢰 구간 생성 
plt.figure(figsize=(12,8))
sns.set(font_scale=1.5)
sns.barplot(data = titanic, x = "pclass", y = "age", hue = "sex", dodge = True)
plt.show()
# --> 겹치는 구간이 없으므로 1등실에 승선한 남성 승객들의 나이가 1등급 승선 여성 승객들보다 많다는 통계적으로 의미있는 결과 도출
```

```python
# pointplot(선형 그래프)
# 통계적인 그래프, 선은 신뢰구간, 선의 가운데는 평균
# 타이타닉 호의 승선 예약을 1000번 반복해서 시뮬레이션 한 결과를 토대로 95%의 신뢰 구간 생성 
plt.figure(figsize=(12,8))
sns.set(font_scale=1.5)
sns.pointplot(data = titanic, x = "pclass", y = "age", hue = "sex", dodge = True)
plt.show()
# --> 1등급 남성 승객들이 여성 승객들보다 나이가 많다, 3등급도 마찬가지, 2등급은 장담할 수 없음
```

#### Jointplots / Regression - 상관 관계를 그래프로 분석 가능

승객들의 나이가 많을수록 더 비싼 티켓을 구매했을까?

승객들의 나이와 지불한 운임이 재난에서 생존하는 데 영향을 미쳤을까?

```python
titanic.head()
```

```python
# jointplot
sns.set(font_scale=1.5)
sns.jointplot(data = titanic, x = "age", y = "fare", height = 8, kind = "scatter")
plt.show()
```

```python
# 육각형이 짙은 색일수록 많은 데이터가 포함
sns.set(font_scale=1.5)
sns.jointplot(data = titanic, x = "age", y = "fare", height = 8, kind = "hex")
plt.show()
```

```python
# 밀도를 나타내는 plot
sns.set(font_scale=1.5)
sns.jointplot(data = titanic, x = "age", y = "fare", height = 8, kind = "kde")
plt.show()
```

```python
# 육각형이 짙은 색일수록 많은 데이터가 포함
sns.set(font_scale=1.5)
sns.jointplot(data = titanic, x = "age", y = "fare", height = 8, kind = "reg")
plt.show()
# 회귀선이 위로 올라가는 양의 상관 관계가 있음
```

```python
sns.set(font_scale=1.5)
sns.lmplot(data = titanic, x = "age", y = "fare", aspect= 1, height=8, hue = "sex")
plt.show()
```

```python
sns.set(font_scale=1.5)
sns.lmplot(data = titanic, x = "age", y = "fare", aspect= 1, height=8, col = "sex") # 칼럼별로 그리기
plt.show()
```

```python
sns.set(font_scale=1.5)
sns.lmplot(data = titanic, x = "age", y = "survived", aspect= 1, height=8,  logistic= True)
plt.show()
```

```python
sns.set(font_scale=1.5)
sns.lmplot(data = titanic, x = "age", y = "survived", aspect= 1, height=8, col = "sex", logistic= True)
plt.show()
# -> 남성은 어릴 수록 생존율이 높고, 여성은 나이가 많을 수록 생존률이 높다.
# -> 남성 승객에서 어린이가 차지하는 비율이 더 높기 때문
```

```python
sns.set(font_scale=1.5)
sns.lmplot(data = titanic, x = "age", y = "survived", aspect= 1, height=8, col = "pclass", logistic= True)
plt.show()
# --> 등급과는 상관없이 나이가 많은 승객일수록 생존율이 낮다.
```

#### Matrixplots / Heatmaps
히트맵을 만들어 표 형식 데이터를 시각화하는 법

```python
titanic.head()
```

```python
pd.crosstab(titanic.sex, titanic.pclass)
```

```python
# annot: 숫자 레이블 표시여부
# fmt: 데이터 타입(정수형-d, 소수점-.2f)
# cmap: 색상 구성
# vmax: 색상 상한치(이 숫자 이상으로 색깔이 짙어지지 않게 표시)
plt.figure(figsize=(12,8))
sns.set(font_scale=1.4)
sns.heatmap(pd.crosstab(titanic.sex, titanic.pclass), annot= True, fmt = "d", cmap = "Reds", vmax = 150)
plt.show()
```

```python
pd.crosstab(titanic.sex, titanic.pclass, values= titanic.survived, aggfunc= "mean")
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.4)
sns.heatmap(pd.crosstab(titanic.sex, titanic.pclass, values= titanic.survived, aggfunc= "mean"), annot= True, cmap = "Reds")
plt.show()
```

```python
titanic.corr()
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.4)
sns.heatmap(titanic.corr(), annot= True, cmap = "Reds")
plt.show()
# -> 1등실을 이용한 승객들이 3등실을 이용한 승객들보다 비싼 요금을 내는 게 당연
# -> parch와 sibsp 사이에 강한 양의 상관관계가 있다(부부가 함께 탑승할 때 아이와 동반한 경우가 많다)
```
