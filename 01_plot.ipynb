### Histograms

- 판다스에서 직접 호출하면 없는 값을 다룰 수 있다
- mat에서 호출하면 null값은 제외하고 그려야한다

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
titanic = pd.read_csv("./Video_Lecture_NBs/titanic.csv")
```

```python
titanic.head()
```

```python
titanic.age.value_counts(bins=10)
```

```python
# kind 파라미터 사용
titanic.age.plot(kind="hist", figsize = (12,8), fontsize = 15,bins = 80)
```

```python
titanic.age.hist(figsize = (12,8),bins = 80, ylabelsize = 15)
plt.show()
```

```python
# mat에서 hist 직접 호출
# 직접 호출 시 bins에 없는 값은 사용할 수 없다(null)
# density 상대 도수로 그리기
# cumulative 누적 합 그래프로 그리기
plt.figure(figsize=(12,8))
plt.hist(titanic.age.dropna(), bins = 80, density=True, cumulative=True)
plt.show()
```

### Scatterplots
산점도를 통해 두 변수간 어떤 상관관계가 있는지 파악할 수 있다

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

```python
titanic = pd.read_csv("./Video_Lecture_NBs/titanic.csv")
```

```python
titanic.head()
```

-가정: 승객의 나이와 표 값은 비례한다

```python
# 2차원 산포도
titanic.plot(kind="scatter", figsize = (15,8), x="age", y="fare", c="r", marker = "o", s = 20)
plt.show()
```

선형으로 나오지 않으므로 비례하지 않는다

```python
# 3차원 산포도
# 생존여부에 따라 다르게 그리기
titanic.plot(kind="scatter", figsize = (15,8), x="age", y="fare", c="survived", marker = "o", s = 20)
plt.show()
```

```python
titanic.plot(kind="scatter", figsize = (15,8), x="age", y="fare", c="pclass", marker = "o", s = 20, colormap = "viridis")
plt.show()
```

- 1등급 클래스가 가장 많이 지불
- age와 pclass간의 상관관계는 크지 않다. 하지만 나이가 어린 승객일 수로 2~3등급 많을 수록 1등급을 이용
